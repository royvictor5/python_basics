```python
import matplotlib.pyplot as plt

xarray = [10, 5,10,40,20]
yarray = [0, 10,30,55,17]

plt.xlabel="Year"
plt.ylabel="Sales"
plt.title="Sales"

plt.plot(xarray,color='r',marker='o',ms =10,mec='r',mfc='g',ls='dashed',lw=4) ## MarkerSize,MarkerEdgeColor,MarkerFaceColor,LineStyle,LineWidth
plt.plot(yarray,color='g',marker='*',ms =10,mec='r',mfc='g',ls='dashed',lw=1.5)

plt.grid(axis='y',lw=0.1,color='m', ls="-" ) ## Don't mention axis parameter to have both
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
colors = [0, 10, 20, 130, 40, 45, 150, 55, 60, 170, 180, 100, 200]
sizes = [20,50,100,200,500,1000,60,90,10,300,600,800,75]

plt.scatter(x, y, s=sizes,c=colors, cmap='viridis')

plt.colorbar()
plt.show()
```
